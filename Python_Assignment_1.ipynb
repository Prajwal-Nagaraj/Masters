```python
'''This Function takes a string that contains some numbers as input and separates the strings and the numbers and returns the sum of numbers'''
def sum_of_digits(s):                                                           #function definition with the variable s as the input
 digits1=[]                                                                     #Instantiation of a list called digits1 to add all the digits to the list
 symbols=[]                                                                     #Instantiation of a list called symbols to add all the non-digits to the list
 z= list(s)                                                                     #converting the input variable into a list
 if (len(z)==0):                                                                #checking if an empty input variable is entered
   print('Empty string Entered!')
 else:
  for i in z:                                                                   #for loop which checks if each element in the list is a digit or not
       if i.isdigit() == 1:                                                     #is.digit() method returns 1 if the element is a digit
        digits1.extend(i)                                                       #add it to the list digits1
        digits1 = list(map(int, digits1))                                       #converting the list elements to integer variable from string variables
       else:                                                                    #if the elements in the list are not digits then
        symbols.extend(i)                                                       #add it the list symbols
       continue

  if (len(digits1)==0):                                                         #if the digits1 list has no digits then 
     print('The sum of digits operation could not detect a digit!')             #print a statement saying there's no digits in the entered variable
     print('The returned input letters are :' , end= '' )                       #and return all the symbols entered by the user separated by ',' symbol
     print([x for x in symbols], sep = ',')                           
  else:                                                                         #if there are digits detected then 
     print('The sum of digits operations performs :', end='')
     print(*[x for x in digits1], sep='+')                                      #print all the digits seprated by the '+' symbol and return the sum of all digit
     print('The extracted non digits are:' , end='')
     print([x for x in symbols], sep = ',')
     print('Output:',sum(digits1))
    


sum_of_digits("10a20")

sum_of_digits("123")

sum_of_digits("we10a20b")

sum_of_digits("united")

sum_of_digits("")
```

```python
# CMT309 - CMT314 2021-2022 Coursework Q2 Test Code
# Oktay Karakus, PhD
# ************************************************************************************
'''
    1) Please copy and pass your codes for pluralize() function below.
    2) Do required changes in function definition for the arguments if needed.
    '''
''' Prajwal Nagaraja(C21097301)
  This Function converts an input string to it's plural form by following a set of rules as explained in the comments''' 



def pluralize(x):                                                               #defining the pluralise function with one variable x
  import re                                                                     #importing regular expressions
  filepath='proper_nouns.txt'                                                   #setting the filepath for the proper nouns folder
  proper_nouns=[line.strip() for line in open(filepath)]                        #opening the file as a list
  f={}                                                                          #Instantiating a dictioanry that will store the result of the operation
  cap=re.findall(r'^[A-Z]',x)                                                   #first we check if the entered input variable is in upper case
  if(len(cap)>0):                                                               #if it is in upper case, we convert it to lower case for easier processing
    x=x.lower()
  pf = re.findall(r's\b',x)                                                     #this expressions checks is the variable is ending with s i.e, it checks if the variable is already in plural form 
  vow = re.findall(r'[aeiou]\b',x)                                              #this expression checks if the variable is ending with a vowel
  con = re.findall(r'[^aeiou]y\b',x)                                            #this expression checks if the variable is ending with a consonant
  fe = re.findall(r'f\b',x)                                                     #This expression checks if the variable is ending with f
  sh = re.findall(r'\w[ch|sh|z]$',x)                                            #This expression checks if the variable is ending with either ch, sh or z

  if (len(x) == 0):                                                             #If an empty string is entered then update the list with the value of f to empty_string
    f.update({'plural':'', 'status':'empty_string!'})                           #and return the dictionary f
    return f
  elif (len(pf)>0):                                                             #If the variable is already in plural form then update the value of f to already_in_plural
    f.update({'plural':x, 'status':'already_in_plural'})                        #and return the dictionary f  
    return f
  elif(x in proper_nouns):                                                      #if the variable entered is a proper noun  then update the value f to proper_noun
    f.update({'plural':x.capitalize(), 'status':'proper_noun'})                 #and return the dictionary f
    return f
  elif(len(vow)>0):                                                             #if the variable entered is ending with a vowel 
    x1 = 's'                                                                    #then append the variable with s then update the value of f
    x = x+ x1                                                                   #and return the dictionary f  
    f.update({'plural':x, 'status':'success'})
    return f
  elif(len(con)>0):                                                             #if the variable entered is ending with a consonant, 
    x2 = list(x.strip(""))                                                      #then remove the last letter and then append the variable 
    x2.pop()                                                                    #with 'ies' and update the value of f        
    x=''.join(x2)                                                               #and return the dictionary f
    x3 = 'ies'
    x= x +x3
    f.update({'plural':x, 'status':'success'})
    return f
  elif(len(fe)>0):                                                              #if the variable entered is ending with f then remove the last letter
    x4=list(x.strip(""))                                                        #and then append the variable with ies
    x4.pop()                                                                    #and return the dictionary f
    x=''.join(x4)
    x5 = 'ves'
    x=x+x5
    f.update({'plural':x, 'status':'success'})
    return f
  elif(len(sh)>0):                                                              #if the varaible entered is ending with either sh, ch or z then append the variable with 'es' 
    x6 = 'es'                                                                   #and return the dictionary f
    x = x+x6
    f.update({'plural':x, 'status':'success'})
    return f
  else:                                                                         #if none of the above conditions are met then just add s to the end of the word
    x7 = 's'                                                                    #and return the dictionary f  
    x=x+x7
    f.update({'plural':x, 'status':'success'})
    return f

### --- IMPORTANT: DO NOT REMOVE OR CHANGE THE CODE BELOW ---
TEST_CASES = """failure
food
Zulma
injury
elf
buzz
computers
PCs

highway
presentation
pouch
adam""".split('\n')

if __name__ == '__main__':

  for test_noun in TEST_CASES:

    print(test_noun,'-->',pluralize(test_noun))
    print('----')

### --- The outputs of test cases are given below. Please compare with your results.
# ************************************************************************************
# failure --> {'plural': 'failures', 'status': 'success'}
# ----
# food --> {'plural': 'foods', 'status': 'success'}
# ----
# Zulma --> {'plural': 'Zulma', 'status': 'proper_noun'}
# ----
# injury --> {'plural': 'injuries', 'status': 'success'}
# ----
# elf --> {'plural': 'elves', 'status': 'success'}
# ----
# buzz --> {'plural': 'buzzes', 'status': 'success'}
# ----
# computers --> {'plural': 'computers', 'status': 'already_in_plural'}
# ----
# PCs --> {'plural': 'PCs', 'status': 'already_in_plural'}
# ----
#  --> {'plural': '', 'status': 'empty_string'}
# ----
# highway --> {'plural': 'highways', 'status': 'success'}
# ----
# presentation --> {'plural': 'presentations', 'status': 'success'}
# ----
# pouch --> {'plural': 'pouches', 'status': 'success'}
# ----
# adam --> {'plural': 'adam', 'status': 'proper_noun'}
# ----
```
